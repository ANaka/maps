```python
import dotenv
from datetime import datetime
from maps.utils import Gmaps
```

```python

config = dotenv.dotenv_values()
gmaps = Gmaps(key=config['api_key'])
```

```python
home_addresses = [
    'Steiner St &, Hayes St, San Francisco, CA 94117',
    '3601 Lyon St, San Francisco, CA 94123',
    ]

workplace_addresses = {
    'One Ferry Building #50, San Francisco, CA 94111',
    'Dolores St &, 19th St, San Francisco, CA 94114', 
}
times= [datetime(2022, 6, 1, 9, 0)]

modes = ['driving', 'walking', 'bicycling', 'transit']

```

```python
results = gmaps.run_queries(
    origin_addresses=home_addresses, 
    destination_addresses=workplace_addresses, 
    times=times, 
    modes=modes)
```

```python
results
```

```python
idx = results.groupby(['origin_address', 'destination_address'])['duration_min'].idxmin()
fastest_route_scores = results.loc[idx]
fastest_route_scores
```

```python

pivot_scores = fastest_route_scores.pivot(index=['origin_address'], columns=['destination_address'], values=['duration_min', 'dist_mi'])
commute_times = pivot_scores.loc[:, 'duration_min']
commute_times
```
